```python

import pandas as pd
dt_housing = pd.read_csv("drive/MyDrive/Colab Notebooks/Housing_2019.csv")
print(dt_housing.head())
```

```python
X = dt_housing.iloc[:,[2,3,5,11]]
print(X)
```

```python
y = dt_housing.price
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 1/3.0) 
```

```python
from sklearn import linear_model
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

```python
y_pred = lm.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
print("Ket qua danh gia theo chi so MSE",round(mse,3)) # = 445099185.233
print("Ket qua danh gia theo chi so RMSE",round(rmse,3)) # = 21097.374
```

```python
# Cau 2
def LR2(X, Y, eta, lanlap, theta0, theta1):
  for k in range(0,lanlap):
    sum0 = 0
    sum1 = 0
    for i in range(0,len(X)):
      h_i = theta0 + theta1*X[i]
      t0 = (Y[i] - h_i)*1
      sum0 = sum0 + t0
      t1 = (Y[i] - h_i)*X[i]
      sum1 = sum1 + t1
    theta0 = theta0 + eta*(sum0/len(X))
    theta1 = theta1 + eta*(sum1/len(X))
  return [round(theta0,3),round(theta1,3)]
```

```python
X_2 = np.array([1,2,4])
Y_2 = np.array([2,3,6])
```

```python
theta = LR2(X_2,Y_2,0.2,2,0,1)
print(theta)
```

```python
# Du bao
XXX = [0,3,5]
for i in XXX:
  YYY = theta[0] + theta[1]*i
  print(round(YYY,3))
```
